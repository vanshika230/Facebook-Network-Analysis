# Visualize graphs with different layouts and options using NetworkX and Python.

1) First, the necessary libraries are imported, including matplotlib, networkx, pandas, and numpy. The code then reads the network data from an Excel file using pd.read_excel() and stores the nodes and edges as dataframes.

2) Next, the graph is created using nx.convert_matrix.from_pandas_edgelist() function by passing the edges dataframe and specifying the source and target columns. The node attributes are then set using nx.set_node_attributes() by passing the nodes dataframe as a dictionary.

3) The first visualization is created using the nx.draw() function, which draws the graph with default options.

4) The second visualization is created using the nx.draw() function again, but this time the node_size option is set to 1 and the node_color option is set to C1, which is a shade of blue.

5) The third visualization uses the nx.spring_layout() function to generate the layout and the pos option is passed to nx.draw() to use this layout. The k option specifies the optimal distance between nodes, which is set to 1.0. The node_color option is set to a range of colors based on the number of nodes.

6) The fourth visualization uses the nx.kamada_kawai_layout() function to generate the layout and the same options as the previous visualization are used.

7) The fifth visualization uses the nx.spiral_layout() function to generate the layout and the same options as the previous visualization are used.

Each visualization is plotted in a new figure with the fig = plt.figure(figsize=(14, 9)) command, which sets the figure size, and the fig.set_facecolor('k') command, which sets the figure background color to black.

```python

# Import necessary libraries
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import numpy as np
from random import sample

# Load data from online Excel file
data_url = 'https://github.com/vanshika230/Facebook-Network-Analysis/blob/main/Social%20Network%20Dataset.xlsx?raw=true'
network_data = pd.read_excel(data_url, sheet_name=['Elements', 'Connections'])
elements_data = network_data['Elements']  # Dataframe of nodes
connections_data = network_data['Connections']  # Dataframe of edges

# Define edge attributes
edge_cols = ['Type', 'Weight', 'When']

# Create a NetworkX graph from the edge dataframe
graph = nx.convert_matrix.from_pandas_edgelist(connections_data,
                                               source='From',
                                               target='To',
                                               edge_attr=edge_cols)

# Create a dictionary of node attributes and add them to the graph
node_dict = elements_data.set_index('Label').to_dict(orient='index')
nx.set_node_attributes(graph, node_dict)
```

```python
nx.draw(graph)
```

```python
fig = plt.figure(figsize = (15, 10))
nx.draw(graph, 
        node_size = 1,
        node_color = 'b')
fig.set_facecolor('k')
```

```python
fig = plt.figure(figsize = (14, 9))
colors = np.linspace(0, 1, len(graph.nodes))
nx.draw(graph, 
        node_size  = 30,
        node_color = colors,
        edge_color = 'b')
fig.set_facecolor('k')
```

```python
colors = np.linspace(0, 1, len(graph.nodes))

k = 1.0
layout = nx.spring_layout(graph, k = k)

fig = plt.figure(figsize = (14, 9))

nx.draw(graph,
        node_size = 30,
        node_color = colors,
        pos = layout, 
        edge_color = 'b')
fig.set_facecolor('k')
```

```python
colors = np.linspace(0, 1, len(graph.nodes))

k = 1.0
layout = nx.kamada_kawai_layout(graph)

fig = plt.figure(figsize=(14, 9))

nx.draw(graph,
        node_size = 30,
        node_color = colors,
        pos = layout, 
        edge_color = 'b')
fig.set_facecolor('k')
```

```python
colors = np.linspace(0, 1, len(graph.nodes))

k = 1.0
layout = nx.spiral_layout(graph)

fig = plt.figure(figsize = (14, 9))

nx.draw(graph,
        node_size = 30,
        node_color = colors,
        pos = layout, 
        edge_color = 'b')
fig.set_facecolor('k')
```
