# Network Data Science with NetworkX and Python

## Add and remove nodes and edges from graphs

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
fig = plt.figure(figsize = (14, 9))

graph = nx.random_internet_as_graph(20)
nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')

fig.set_facecolor('k')
```

## Adding a edge

```python
fig = plt.figure(figsize = (14, 9))

graph.add_edge(18, 13)
nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')

fig.set_facecolor('k')
```

## Adding a Node

```python
fig = plt.figure(figsize = (14, 9))

graph.add_node(40)
nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')

fig.set_facecolor('k')
```

## Adding new nodes and edges

```python
fig = plt.figure(figsize = (14, 9))

graph = nx.Graph()
new_nodes = ['q', 'w', 'e', 'r', 't', 'y']
new_edges = [('q', 'w'), ('e', 'r'),  ('t', 'y'),
             ('q', 'y'), ('w', 't')]
graph.add_nodes_from(new_nodes)
graph.add_edges_from(new_edges)

nx.draw(graph, 
        node_color = 'C1',
        with_labels = True,
        edge_color = 'g')

fig.set_facecolor('k')
```
